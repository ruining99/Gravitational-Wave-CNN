```python
from ipynb.fs.full.False_signals_functions import damp_sin
import pycbc.noise
import pycbc.psd
import pycbc.filter
import pylab
from pycbc.filter import sigma
from pycbc.waveform import get_td_waveform
from pycbc.types.timeseries import load_timeseries
import matplotlib.pyplot as plt
```

```python
# Generate psd
flow = 20.0
delta_f = 1.0 / 16
flen = int(2048 / delta_f) + 1
psd = pycbc.psd.aLIGOZeroDetHighPower(flen, delta_f, flow)

# Generate 1 seconds of noise at 4096 Hz, randomized for each trial.
delta_t = 1.0 / 4096
tsamples = int(1.0 / delta_t)
ts = pycbc.noise.noise_from_psd(tsamples, delta_t, psd)

# plot
pylab.plot(ts.sample_times, ts)
pylab.show()

print(ts.shape)
```
